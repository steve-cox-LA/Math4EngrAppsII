# 6. Taylor's Formula


## Every function is a polynomial 

If one deritave reveals the local slope of a function we may naturally ask about the slope of the slope. This is encoded in the second derivative

$$
  f''(x) = \lim_{h\to 0}\frac{f'(x+h)-f'(x)}{h}.
$$

Given all of our great rules for derivatives we really don't need to keep on taking these challenging limits. 

In particular as the derivative of $\sin$ is $\cos$ and the derivative of $\cos$ is $-\sin$ we see that 

**the second derivative of $\sin$ is $-\sin$**. 

By the exact same reasoning we also deduce that

**the second derivative of $\cos$ is $-\cos$**. 

Continuing on, please convince yourself that the third derivative of $\sin$ is $-\cos$ while the third derivative of $\cos$ is $\sin$. What about fourth derivatives and so on?

**Taylor's Theorem** states that the value of a function, $f$, near a point, $x$, can be approximated by a polynomial whose coefficients are derivatives of $f$ at $x$. More precisely, Taylor sets up the problem

$$
  f(x+h) = f(x) + ah + bh^2 + ch^3 + dh^4 + \hbox{multiples of higher powers of }h  \tag{Taylor}
$$

and solves for $a$, $b$, $c$, and so on. By a slight switch in point of view we note that taking the derivative of Eq. Taylor
with respect to $h$ delivers

$$
  f'(x+h) = a + 2bh + 3ch^2 + 4dh^3 + \hbox{multiples of higher powers of }h  \tag{Taylor'}
$$

If we then evaluate both sides of Taylor' as $h\to 0$, we discover that 

$$
  f'(x) = a.
$$

Continuing in this fashion, we take the derivative of Eq. Taylor'
with respect to $h$  and find

$$
  f''(x+h) = 2b + 6ch + 12dh^2 + \hbox{multiples of higher powers of }h  \tag{Taylor''}
$$

If we then evaluate both sides of Taylor'' as $h\to 0$, we discover that 

$$
  f''(x) = 2b.
$$

Continuing in this fashion, we take the derivative of Eq. Taylor''
with respect to $h$  and find

$$
  f'''(x+h) = 6c + 24dh + \hbox{multiples of higher powers of }h  \tag{Taylor'''}
$$

If we then evaluate both sides of Taylor''' as $h\to 0$, we discover that 

$$
  f'''(x) = 6c.
$$

Continuing on you will not be suprised to find that

$$
  f''''(x) = 24d = 4!d
$$

where the ! symbol denotes the factorial function, $4!=4\cdot 3\cdot 2\cdot 1$.

As a result, we have arrived at Taylor's Formula

$$
  \boxed{ f(x+h) = f(x) + f'(x)h + \frac{f''(x)}{2!}h^2 + \frac{f'''(x)}{3!}h^3 + \frac{f''''(x)}{4!}h^4 + \cdots}
    \tag{TAYLOR}
$$

This will make much more sense when we work out detailed examples.

## Taylor's Formula for sine

We choose $f=\sin$ and $x=0$ and so (TAYLOR) delivers

$$
  \sin(h) = \sin(0) + \sin'(0)h + \frac{\sin''(0)}{2!}h^2 + \frac{\sin'''(0)}{3!}h^3 + \frac{\sin''''(0)}{4!}h^4 + \cdots \\
          = h - \frac{h^3}{3!} + \frac{h^5}{5!} - \frac{h^7}{7!} + \cdots \tag{T-sin}
$$

because $\sin(0)=0$ and an even number of derivatives of $\sin$ at 0 is proportional to $\sin(0)$, while $\cos(0)=1$ and
an odd number of derivatives of $\sin$ at 0 is proportional to $\cos(0)$.
 

```python
# Build and plot the Taylor Approximations of sin
import numpy as np
import matplotlib.pyplot as plt
import math

h = np.linspace(-np.pi, np.pi, 100)    # 100 h points  
plt.plot(h, np.sin(h),'k')             # plot sin(h) vs. h

Tsin = h                               # build and plot the 1st term in the Taylor Series
plt.plot(h, Tsin, 'r--')

Tsin = Tsin - h**3/math.factorial(3)   # add on the next term in the Taylor Series and plot
plt.plot(h, Tsin, 'b--')

Tsin = Tsin + h**5/math.factorial(5)   # add on the next term in the Taylor Series and plot
plt.plot(h, Tsin, 'g--')

Tsin = Tsin - h**7/math.factorial(7)   # add on the next term in the Taylor Series and plot
plt.plot(h, Tsin, 'c--')

plt.grid('on')
plt.legend(['$y=\sin(h)$','$y=h$','$y=h-h^3/3!$','$y=h-h^3/3!+h^5/5!$', '$y=h-h^3/3!+h^5/5!-h^7/7!$'],loc=(1.04,0))
plt.xlabel('h')
plt.ylabel('y')
plt.ylim([-2,2])
plt.title('Exploring the Tayor Series of sine');
```

Can you enjoy the beautiful Taylor evolution? With an eye on $h>0$ we see that black is truth, red overshoots, then blue undershoots, then green overshoots, then cyan undershoots. Each stage gets closer and closer until by the time we reach order 7 we are super close to nailing $\sin$ over its **whole** domain.
