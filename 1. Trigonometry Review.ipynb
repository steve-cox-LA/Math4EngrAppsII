# 1. Trigonometry Review

Let us review the key definitions of trigonometry. We will measure angles in radians. In the picture below, $\theta$ is the angle, $r$ is the radius, and $s$ is the length of the arc. This picture illustrates the beautiful relationship

$$
   \boxed{\theta = \frac{s}{r}}  \tag{1}
$$

As a ratio of lengths, $\theta$ has NO dimensions.

```python
import numpy as np                  # import all the numerical math methods
import matplotlib.pyplot as plt     # import all the plotting methods

t = np.linspace(0, np.pi/4, 100)   # make 100 evenly spaced points between 0 and pi/4
plt.plot(np.cos(t), np.sin(t))     # plot an arc of the unit circle (in blue, the default color)

theta = np.pi/4

plt.plot([0, 1],[0, 0],'k')                                                 # plot the black line
plt.text(0.5*np.cos(1.1*theta), 0.5*np.sin(1.1*theta),'r',fontsize=20)
plt.plot([0, np.cos(theta)],[0, np.sin(theta)],'r')                         # plot the red ray
plt.text(1.05*np.cos(theta/2), 1.05*np.sin(theta/2),'s',fontsize=20)
plt.text(0.25*np.cos(theta/2), 0.25*np.sin(theta/2),'$\\theta=s/r$',fontsize=20)
plt.text(0,0.65,'Figure 1',fontsize=20)
plt.axis('off')
plt.axis('equal');
```

You may recall that arc length of a full circle (often called **circumference** or perimeter) of radius $r$ is $2\pi r$. Rather than simply stating this formula, we will be able to derive it from simpler facts.

We next associate two important functions with each angle. Both functions, as with the angle itself, are ratios of (signed) lengths. These lengths come from dropping a green vertical onto the picture above, arriving at

```python
plt.plot([0, np.cos(theta)],[0, 0],'k')  
plt.text(0.5*np.cos(1.1*theta), 0.5*np.sin(1.1*theta),'r',fontsize=20)
plt.plot([0, np.cos(theta)],[0, np.sin(theta)],'r')
plt.plot([np.cos(theta), np.cos(theta)],[0, np.sin(theta)],'g')
plt.text(1.05*np.cos(theta), 0.5*np.sin(theta),'$r*\sin(\\theta)$',fontsize=20)
plt.text(0.35*np.cos(theta), -.1,'$r*\cos(\\theta)$',fontsize=20)
plt.text(0.25*np.cos(theta/2), 0.25*np.sin(theta/2),'$\\theta$',fontsize=20)
plt.text(0,0.65,'Figure 2',fontsize=20)
plt.axis('off')
plt.axis('equal');
```

You may also know this picture in wordier terms, like those below

```python
plt.plot([0, np.cos(theta)],[0, 0],'k') 
plt.text(0.5*np.cos(1.1*theta)-.2, 0.5*np.sin(1.1*theta),'hyp',fontsize=20)
plt.plot([0, np.cos(theta)],[0, np.sin(theta)],'r')
plt.plot([np.cos(theta), np.cos(theta)],[0, np.sin(theta)],'g')
plt.text(1.05*np.cos(theta), 0.5*np.sin(theta),'opp',fontsize=20)
plt.text(0.35*np.cos(theta), -.1,'adj',fontsize=20)
plt.text(0.25*np.cos(theta/2), 0.25*np.sin(theta/2),'$\\theta$',fontsize=20)
plt.text(0,0.65,'Figure 3',fontsize=20)
plt.axis('off')
plt.axis('equal');
```

where hyp is short for hypotenuse, opp for opposite and adj for adjacent. These are each lengths and we define our trig functions via

$$
  \cos(\theta) = \frac{\hbox{adj}}{\hbox{hyp}} \hskip 0.25in\hbox{and}\hskip 0.25in \sin(\theta) = \frac{\hbox{opp}}{\hbox{hyp}}
$$

As our angle $\theta$ moves from $0$ to $2\pi$ these functions trace out the beatiful curves below

```python
theta = np.linspace(0, 2*np.pi, 1000)  
plt.plot(theta, np.cos(theta), 'r', linewidth = 4)
plt.plot(theta, np.sin(theta), 'b', linewidth = 4)
plt.legend(['cos','sin'])
plt.title('Figure 4',fontsize=20)
plt.grid('on');
```

Our motivation for studying these functions comes from Joseph Fourier, who argued that $\sin$ and $\cos$ are the building blocks of every function in the world. As engineers we are interested in what we can build with these blocks. So first, of course, we must know these blocks cold!

We will review a few of their elementary properties before moving onto the more challenging identities. From the graph above we see

$$
  \cos(0) = 1, \hskip 0.25in \sin(0) = 0 \\
  \cos(\pi/2) = 0, \hskip 0.25in \sin(\pi/2) = 1 \\
  \cos(\pi) = -1, \hskip 0.25in \sin(\pi) = 0 \\
  \cos(3\pi/2) = 0, \hskip 0.25in \sin(3\pi/2) = -1
$$

I wish to emphasize that these are not things we memorize, but things we **see**.

With Pythagoreas we have also learned to see that the sum of square of the lengths of the legs of right triangle is 
the square of the length of the hypotenuse. With regard to our second figure above, this reads

$$
  r^2\cos^2(\theta) + r^2\sin^2(\theta) = r^2.
$$

On dividing throught by $r^2$ we arrive at our first awesome identity

$$
  \boxed{\cos^2(\theta) + \sin^2(\theta) = 1}  \tag{2}
$$

and note that this holds for **every** angle $\theta$.

So far we have followed the convention of measuring angles as positive as we move counterclockwise from the positive $x$ axis. If we reverse our direction it makes sense to reverse the sign of our angle. If we reflect our second figure over the $x$-axis we find

```python
# solve exercise 2 here, borrow code from above

theta = np.linspace(0, 2*np.pi, 1000)  
plt.plot(theta, np.cos(theta)**2 + np.sin(theta)**2, 'r', linewidth = 4)
plt.xlabel('$\\theta$', fontsize=12)
plt.ylabel('$\cos^2(\\theta) + \sin^2(\\theta)$', fontsize=12)

plt.title('Exercise 2',fontsize=20)
plt.grid('on');
```

```python
theta = np.pi/4
plt.plot([0, np.cos(theta)],[0, 0],'k')   
plt.text(0.5*np.cos(-1.1*theta)-0.1, 0.5*np.sin(-1.1*theta),'r',fontsize=20)
plt.plot([0, np.cos(-theta)],[0, np.sin(-theta)],'r')
plt.plot([np.cos(-theta), np.cos(-theta)],[0, np.sin(-theta)],'g')
plt.text(1.05*np.cos(-theta), 0.5*np.sin(-theta),'$r*\sin(-\\theta)$',fontsize=20)
plt.text(0.35*np.cos(-theta), 0.1,'$r*\cos(-\\theta)$',fontsize=20)
plt.text(0.25*np.cos(-theta/2), 0.25*np.sin(-theta/2),'$-\\theta$',fontsize=20)
plt.text(0,-0.65,'Figure 5',fontsize=20)
plt.axis('off')
plt.axis('equal');
```

Comparing the original and reflected triangles we see that the **adjacent leg** remains unchanged, and so

$$
  \boxed{\cos(-\theta) = \cos(\theta)} \tag{3}
$$

while the **opposite leg** is now in the land of negative $y$ and so

$$
  \boxed{\sin(-\theta) = -\sin(\theta)}  \tag{4}
$$

Let us confirm that python knows these facts.

```python
theta = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(theta, np.cos(theta), 'r')
plt.plot(theta, np.sin(theta), 'b')
plt.legend(['cos','sin'])
plt.title('Figure 6', fontsize=20)
plt.grid('on');
```

Now that we can handle negative angles let's next make sure we understand that our words **opposite** and **adjacent** are relative to a chosen angle. Let's first calculate the third angle in our second figure by completing this figure to a rectangle, all of whose angles must be $\pi/2$.

```python
theta = np.pi/4
plt.plot([0, np.cos(theta)],[0, 0],'k') 
plt.plot([0, np.cos(theta)],[np.sin(theta), np.sin(theta)],'k--') 
plt.text(0.5*np.cos(1.1*theta), 0.5*np.sin(1.1*theta),'r',fontsize=20)
plt.plot([0, np.cos(theta)],[0, np.sin(theta)],'r')
plt.plot([np.cos(theta), np.cos(theta)],[0, np.sin(theta)],'g')
plt.plot([0,0],[0, np.sin(theta)],'g--')
plt.text(1.05*np.cos(theta), 0.5*np.sin(theta),'$r*\sin(\\theta)$',fontsize=20)
plt.text(0.35*np.cos(theta), -.1,'$r*\cos(\\theta)$',fontsize=20)
plt.text(0.25*np.cos(theta/2), 0.25*np.sin(theta/2),'$\\theta$',fontsize=12)
plt.text(0.25*np.cos(3*theta/2)-0.075, 0.25*np.sin(3*theta/2),'$\\pi/2-\\theta$',fontsize=12)
plt.text(0.85*np.cos(theta)-.075, 0.85*np.sin(theta),'$\\theta$',fontsize=12)
plt.text(0.85*np.cos(theta)-.075, 0.85*np.sin(theta)-.1,'$\\pi/2-\\theta$',fontsize=12)
plt.text(0,0.65,'Figure 7',fontsize=20)
plt.axis('off')
plt.axis('equal');
```

Now do you **see** that the opposite to $\theta$ is the adjacent to $\pi/2-\theta$, and vice versa? If so, then you can **see**

$$
  \boxed{\cos(\theta)=\sin(\pi/2-\theta) \hskip 0.25in\hbox{and}\hskip 0.25in \sin(\theta) = \cos(\pi/2-\theta)}  \tag{5}
$$
