# 5. Derivatives Galore

We will need to be able to differentiate combinations of algebraic and trigonometric functions. Given a general function, $f(x)$, we pursue its derivative

$$
  f'(x) \equiv \lim_{h\to 0}\frac{f(x+h)-f(x)}{h}.
$$

# The Power Rule

## $f(x) = c$

In this case $f$ is constant and so (flat roofs have slope 0)

$$
  f'(x) \equiv  \lim_{h\to 0}\frac{c-c}{h} = \lim_{h\to 0}\frac{0}{h} = 0
$$

and we see that **the derivative of a constant function is 0.**

## $f(x) = x$

In this case 

$$
  f'(x) \equiv \lim_{h\to 0}\frac{x+h-x}{h} = \lim_{h\to 0}\frac{h}{h} = \lim_{h\to 0}1 = 1
$$

and we see that **the derivative of x is 1.**

## $f(x) = 5x$

In this case

$$
  f'(x) \equiv \lim_{h\to 0}\frac{5(x+h)-5x}{h} = \lim_{h\to 0}\frac{5h}{h} = \lim_{h\to 0}5 = 5
$$

and we see that **the derivative of 5x is 5.** Of course there is nothing special about 5. You recognize that
**the derivative of mx is m for any number m.**

## $f(x) = ax^2$

In this case 

$$
  f'(x) \equiv \lim_{h\to 0}\frac{a(x+h)^2-ax^2}{h} = \lim_{h\to 0}\frac{2axh+ah^2}{h} = \lim_{h\to 0}(2ax + ah) = 2ax
$$

and we see that **the derivative of $ax^2$ is $2ax$ for any number $a$.**

## $f(x) = ax^{-1} = \frac{a}{x}$

In this case 
$$
  f'(x) \equiv \lim_{h\to 0}\frac{1}{h}\left(\frac{a}{x+h}-\frac{a}{x}\right) 
        = \lim_{h\to 0}\frac{a}{h}\frac{x-(x+h)}{x(x+h)} = \lim_{h\to 0}\frac{-a}{x(x+h)} = \frac{-a}{x^2}
$$

and we see that **the derivative of $ax^{-1}$ is $-ax^{-2}$ for any number $a$.**

# The Product Rule

We will often need to differentiate products like $x^2\sin(x)$ and so, in general, we must work out

$$
  (fg)'(x) \equiv \lim_{h\to 0}\frac{f(x+h)g(x+h)-f(x)g(x)}{h}
$$

The key here is to add and subtract terms from the numerator in order to break it into pieces that look like derivatives of $f$ and $g$. To wit using <span style='color:Red'> red  </span> to denote the new stuff,

$$
  (fg)'(x) \equiv \lim_{h\to 0}\frac{f(x+h)g(x+h)-f(x)g(x) {\color{red}{ + f(x+h)g(x)-f(x+h)g(x)}}}{h} \\
     = \lim_{h\to 0}\left(f(x+h)\frac{g(x+h)-g(x)}{h} + g(x)\frac{f(x+h)-f(x)}{h}\right) \\
     = f(x)g'(x) + g(x)f'(x).
$$

With $f(x)=x^2$ and $g(x)=\sin(x)$ we see that the derivative of their product is

$$
  x^2\cos(x) + 2x\sin(x)
$$

# The Quotient Rule

This same strategy will help us differentiate quotients like $\sin(x)/\cos(x)$. In we must work out

$$
  (f/g)'(x) \equiv \lim_{h\to 0}\frac{1}{h}\left(\frac{f(x+h)}{g(x+h)}-\frac{f(x)}{g(x)}\right)
    = \lim_{h\to 0}\frac{1}{h}\frac{g(x)f(x+h)-g(x+h)f(x)}{g(x)g(x+h)}
$$

We again add and subtract terms from the numerator in order to break it into pieces that look like derivatives of $f$ and $g$. To wit,

$$
  (f/g)'(x) \equiv \lim_{h\to 0}\frac{1}{h}\frac{g(x)f(x+h)-g(x+h)f(x) {\color{red}{ + g(x)f(x) - g(x)f(x)}}}{g(x)g(x+h)} \\
     = \lim_{h\to 0}\left(\frac{1}{g(x+h)}\frac{f(x+h)-f(x)}{h} - \frac{f(x)}{g(x)g(x+h)}\frac{g(x+h)-g(x)}{h}\right) \\
     = \frac{f'(x)}{g(x)} - \frac{f(x)g'(x)}{g^2(x)} \\
     = \frac{f'(x)g(x) - f(x)g'(x)}{g^2(x)}.
$$

Hence, with $f(x)=\sin(x)$ and $g(x) = \cos(x)$ we find that

$$
  \tan'(x) = \frac{\cos(x)\cos(x) - \sin(x)(-\sin(x))}{\cos^2(x)} = \frac{1}{\cos^2(x)}.
$$

# The Chain Rule

Finally, we must consider compositions like $\sin(5x)$, or in general

$$
  (f(g))'(x) \equiv \lim_{h\to 0} \frac{f(g(x+h))-f(g(x))}{h}
$$

Now here instead of adding 0 we multiply by 1, to wit

$$
   (f(g))'(x) = \lim_{h\to 0} \frac{f(g(x+h))-f(g(x))}{h}{\color{red}{\frac{g(x+h)-g(x)}{g(x+h)-g(x)}}} \\
   = \lim_{h\to 0} \frac{f(g(x+h))-f(g(x))}{g(x+h)-g(x)}\frac{g(x+h)-g(x)}{h} \\
   = f'(g(x))g'(x)
$$

And so, with $f(x)=\sin(x)$ and $g(x)=5x$ we find

$$
  (\sin(5x))' = 5\cos(5x).
$$

Similarly, with $f(x) = x^2$ and $g(x) = \sin(x)$ we find 

$$
  (\sin^2(x))' = f'(g(x))g'(x) = 2\sin(x)\cos(x)
$$
